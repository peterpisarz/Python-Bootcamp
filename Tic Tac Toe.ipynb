```python
# Run using jupiter notebook to use IPython.display functionality
# Function to establish the initial board game setup
from IPython.display import clear_output

def display_board(board):
    clear_output()
    print(' ' + board[7]+ ' | ' +board[8]+ ' | ' + board[9])
    print('-----------')
    print(' ' + board[4]+ ' | ' +board[5]+ ' | ' + board[6])
    print('-----------')
    print(' ' + board[1]+ ' | ' +board[2]+ ' | ' + board[3])

# Function to select which Player will be X or O
def player_input():
    marker = " "
    
    while not (marker == "X" or marker == 'O'):
        marker = input("Player 1, would you like to be X or O? ").upper()
        if marker == "X":
            print("Ok player 1 you're X and player 2 is O")
            return ("X", "O")
        elif marker == "O":
            print("Ok player 1 you're O and player 2 is X")
            return("O", "X")
        else:
            print("Sorry, only chose X or O")
            marker = " "

# Function which places the player's marker at the desired location
def place_marker(board, marker, position):
    board[position] = marker

# Function which establishes the criterea for winning a match
def win_check(board, mark):
    
    return ((board[7] == mark and board[8] == mark and board[9] == mark) or # across the top
    (board[4] == mark and board[5] == mark and board[6] == mark) or # across the middle
    (board[1] == mark and board[2] == mark and board[3] == mark) or # across the bottom
    (board[7] == mark and board[4] == mark and board[1] == mark) or # down the left side
    (board[8] == mark and board[5] == mark and board[2] == mark) or # down the middle
    (board[9] == mark and board[6] == mark and board[3] == mark) or # down the right side
    (board[7] == mark and board[5] == mark and board[3] == mark) or # diagonal
    (board[9] == mark and board[5] == mark and board[1] == mark)) # diagonal

# Function deciding which player goes first using the random module
import random

def choose_first():
    starter = random.randrange(0,2)
    if starter == 0:
        return "Player 1"
    else:
        return "Player 2"

# Function that checks if a space on the board is free
def space_check(board, position):
    return board[position] == " "

# Function that checks if the board is full without a victory
def full_board_check(board):
    for n in range(1,10):
        if space_check(board,n):
            return False
    return True

# Function which asks for the player's next position as a number 1-9
# Layout of the board based on numerical keypad
# Includes a nested function check to see if the space is free using space_check
def player_choice(board):
    choice = 0
    while choice == 0:
        # Added error exception incase non integers are entered
        try:
            choice = int(input("Pick a position 1-9 "))
        except:
            print("Not an acceptable input! Pick a position 1-9 ")
            choice = 0
            continue     

        if choice in range(1,10) and space_check(board, choice):
            return choice
        elif choice not in range(1,10):
            print("That is out of range. Pick a position 1-9")
            choice = 0
        else:
            print("That space is not free")
            choice = 0

# A function asking if players would like to replay after the game is finished
def replay():
    playagain = ""
    playagain = input("Do you want to play again? Type Y or N ").upper()
    if playagain == "Y":
        return True
    elif playagain == "N":
        return False

# Main code for game play making using of the above defined functions
print('Welcome to Tic Tac Toe!\n')

#while True:
while True:
    # Set the game up here
    theboard=[' ']*10
    player1, player2 = player_input()
    turn = choose_first()
    print(turn + " will go first!")
    
    play_game = input("Ready to play? Y or N: ")
    if play_game.lower()[0] == 'y':
        game_on = True
    else:
        game_on = False
    
    while game_on == True:
        if turn == 'Player 1':
            display_board(theboard)
            choice = player_choice(theboard)
            place_marker(theboard,player1,choice)
            if win_check(theboard, player1):
                display_board(theboard)
                print("Congratulations " + player1 + ", you won the game")
                game_on = False
            else:
                if full_board_check(theboard):
                    print("You both suck, Cats game!")
                    game_on = False
                else:
                    turn = "Player 2"
        else:
            display_board(theboard)
            choice = player_choice(theboard)
            place_marker(theboard,player2,choice)
            if win_check(theboard, player2):
                display_board(theboard)
                print("Congratulations " + player2 + ", you won the game")
                game_on = False
            else:
                if full_board_check(theboard):
                    print("You both suck, Cats game!")
                    game_on = False
                else:
                    turn = "Player 1"
    if not replay():
        print("Ok, thanks for playing!")
        break
```
